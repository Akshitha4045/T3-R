```python
import pandas as pd
```

```python
df=pd.read_csv("/Rossmann Stores Data.csv")
```

```python
df
```

```python
# prompt: from the above data in the file give me linear regression graph code between the sales and customers in python and also represent the title in that graph

import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot with regression line
sns.regplot(x='Customers', y='Sales', data=df)

# Set the title of the plot
plt.title('Linear Regression between Sales and Customers')

# Display the plot
plt.show()
```
